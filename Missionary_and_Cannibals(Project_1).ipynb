# Problem statement
Let's create famous 3 missionaries 3 cannibals crossing the river game. Rules are as follows.<br>
[Demonstration](https://www.novelgames.com/en/missionaries/).

## Rules

1. Number of `missionaries` on either side should never be less than number of `cannibals`
2. Only one or 2 people can travel through boat at a time.
3. You win the game when all `missionaries` and `cannibals` on the right go to the left side



## Steps to solve the problem

### Step 1
- Let's display number of `cannibals(C)` and `missionaries(M)` on the left and right.
- Let's display `boat(B)` on the appropriate side starting from the right.

Initial Display :-  
`M = 0 C = 0  |-----B|  M = 3 C = 3`
<br>

### Step 2
Let's accept number of `missionaries` and then number of `cannibals`.
Example: Initially if 1 `missionaries` and 1 `cannibals` goes from right to left the game display will look like this

`M = 1 C = 1  |B-----|  M = 2 C = 2`

(Let's keep the side of the boat in mind)

Let's don't forget to add and subtract number of missonary on both sides accordingly

### Step 3
Let's define Invalid moves and add conditional `'if'` statements accordingly

**Rules**
1. Number of people in a Boat should be 1 or 2
2. Number of chosen `missionaries` or `cannibals` should not be greater than number of available `missionaries` or `cannibals` on the given side.
   In above step one should not be able subtract 2 missionaries from left as only 1 is available.
3. An additional condition for you to figure out!

Defining variables initially:

```python
boat_side = "Right"
missionaries_on_right = 3
cannibals_on_right = 3
missionaries_on_left = 0
cannibals_on_left = 0
```

```python
missionaries = int(input("Enter Number of Missionaries..."))
cannibals = int(input("Enter Number of Cannibals..."))
```

Else print "Invalid move"

```python
if (missionaries + cannibals) != 1 and (missionaries + cannibals) != 2:
    print("Invalid Move")
```

Adding if else statement to check the value of `boat_side` variable

```python
if boat_side == "Right":
    print("Boat is on right side....")
else:
    print("Boat is on Left side......")
```

### Flow of the combined code

if boat_side == "Right":
    Inside the scope of if statement for right side
    
elif boat_side == "Left":
    Inside the scope of if statement for left side    


```python
if boat_side == 'Right':
  if missionaries_on_right < missionaries or cannibals_on_right < cannibals:
      print("Invalid Move")
  missionaries_on_right -= missionaries
  cannibals_on_right -= cannibals
  missionaries_on_left += missionaries
  cannibals_on_left += cannibals
  print('M =', missionaries_on_left, 'C =', cannibals_on_left, '|B_____|', 'M =', missionaries_on_right, 'C =', cannibals_on_right)
```

```python
if boat_side == 'Left':
  if missionaries_on_left < missionaries or cannibals_on_left < cannibals:
      print("Invalid Move")
  missionaries_on_left -= missionaries
  cannibals_on_left -= cannibals
  missionaries_on_right += missionaries
  cannibals_on_right += cannibals
  print('M =', missionaries_on_left, 'C =', cannibals_on_left, '|_____B|', 'M =', missionaries_on_right, 'C =', cannibals_on_right)
```

Now, Let's add conditions to check if the game is over:
1. If number of missionaries is less that number of cannibals on right side and
   number of missionaries are greater than zero.
2. If number of missionaries is less that number of cannibals on left side and
   number of missionaries are greater than zero.
3. If over print 'YOU LOSE'


```python
if missionaries_on_right < cannibals_on_right and missionaries > 0:
  print("YOU LOSE")
if missionaries_on_left < cannibals_on_left and missionaries > 0:
  print("YOU LOSE")
```

Let's add conditions to check if you have won the game:
1. Number of missionaries and cannibals on the left side should be equal to 3.
2. If condition matches print 'YOU WIN'.

```python
if missionaries_on_left and cannibals_on_left == 3:
  print("YOU WON")
```

The game must go on until you win or lose so lets put all conditions inside the infinite loop
1. Now for all conditions make appropriate changes.
2. If a move is invalid then without making any changes in the board you have to 'continue'.
   So make changes in all the conditions.
3. Now, if you Win or Lose the game 'break' the loop and get out of it.
4. After getting out of the loop print "Game Over".

```python
boat_side = "Right"
missionaries_on_right = 3
cannibals_on_right = 3
missionaries_on_left = 0
cannibals_on_left = 0

print('\U0001f479' * missionaries_on_left,'\U0001f482' * cannibals_on_left,'\U0001f30a''\U0001f30a''\U0001f30a''\U0001f30a''\U0001f30a''\U0001f6A2','\U0001f479' * missionaries_on_right,'\U0001f482' * cannibals_on_right)

while True:

  missionaries = int(input("Enter Number of Missionaries or enter 10 to quit"))
  if missionaries == 10:
    print('YOU QUIT')
    break

  cannibals = int(input("Enter Number of Cannibals..."))

  if (missionaries + cannibals) != 1 and (missionaries + cannibals) != 2:
      print("Invalid Move")
      continue

  if boat_side == 'Right':

    if missionaries_on_right < missionaries or cannibals_on_right < cannibals:
        print("Invalid Move")
        continue

    missionaries_on_right -= missionaries
    cannibals_on_right -= cannibals

    missionaries_on_left += missionaries
    cannibals_on_left += cannibals

    print('\U0001f479' * missionaries_on_left,'\U0001f482' * cannibals_on_left,'\U0001f6A2''\U0001f30a''\U0001f30a''\U0001f30a''\U0001f30a''\U0001f30a','\U0001f479' * missionaries_on_right,'\U0001f482' * cannibals_on_right)

    boat_side = 'Left'

  elif boat_side == 'Left':
    if missionaries_on_left < missionaries or cannibals_on_left < cannibals:
        print("Invalid Move")
        continue

    missionaries_on_left -= missionaries
    cannibals_on_left -= cannibals

    missionaries_on_right += missionaries
    cannibals_on_right += cannibals

    print('\U0001f479' * missionaries_on_left,'\U0001f482' * cannibals_on_left,'\U0001f30a''\U0001f30a''\U0001f30a''\U0001f30a''\U0001f30a' '\U0001f6A2','\U0001f479' * missionaries_on_right,'\U0001f482' * cannibals_on_right)
    boat_side = 'Right'

  if missionaries_on_right < cannibals_on_right and missionaries > 0:
    print("YOU LOSE")
    break
  elif missionaries_on_left < cannibals_on_left and missionaries > 0:
    print("YOU LOSE")
    break

  if missionaries_on_left and cannibals_on_left == 3:
    print("YOU WON")
    break

print("GAME OVER!!!")
```
