## Fermi Pico Bagel
Let's try playing game and understand the rules : https://communicrossings.com/html/js/pfb.htm

## Step 1
Let's store any number in string format(number to be guessed by player) in a variable `original_number`.

For Example: original_number = '123'


```python
original_number = '123'
```

## Step 2
Let's take a number as user input and save it in a variable `guess_number`.(Make sure that the datatype of guess_number is same as of original number i.e- `str`)

```python
guess_number = input('Enter a number: ')
```

## Step 3
Let's check if total number of digits in `original_number` and `guess_number` is not same(Use `len()` function) then ask user for valid input.(First use `len()` to get total digits in `original_number` and `guess_number`.)

```python
if len(original_number) != len(guess_number):
  guess_number = input(f'Enter {len(original_number)} digit number:')
```

## Step 4
Using `set()`, Let's check if the digits are repeated in a `guess_number` .
Note that datatype `set` allows unique values only.

For ex. if `user_input`is 122, then it should print the string 'Duplicate number'  as digit 2 is repeated.

```python
if(len(guess_number) != len(set(guess_number))):
  print('Duplicate numbers are not allowed!')
```

## Step 5
Let's check winning condition i.e. if both `guess_number` and `original_number` are same, then print 'Fermi' as many number of times as number of digits in a number and print ' You win !! '.


```python
if int(original_number) - int(guess_number) == 0:
  print('Fermi' * len(original_number))
  print('YOU WON!!!')
```

## Step 6
Let's create empty list called `output` where we will store the output in the form of list.

For example: output = [ ' Fermi ', ' Pico ' ,' Pico ' ] .

```python
output = []
```

## Step 7
Let's check if any digit and it's position in both `guess_number` and `original_number` are same then append 'Fermi' to `output` list or if only digit matches and not the position then append 'Pico' to `output`.


```python
for i in range(len(original_number)):
  for j in range(len(guess_number)):
    if original_number[i] == guess_number[j]:
      if i == j:
        output.append('Fermi')
      else:
        output.append('Pico')
```

## Step 8
 Let's define a variable called `output_string` which has all values from a list `output` with each value separated by space. (Use 'for' loop to get all the values from a list `output`.)

```python
output_string = ''
for items in output:
  output_string = output_string + ' ' + items
print(output_string)
```

## Step 9
Let's check if neither any digit nor any position matches i.e. if length of `output` list is zero then print `Bagel`.

Otherwise print `output_string`.

```python
if len(output) == 0:
  print('Bagel')
else:
  print(output_string)
```

## Complete game

```python
import random

original_number = str(random.randint(1, 1000))

while True:
  guess_number = input('Enter a number between 1 and 1000:')
  output = []

  if len(original_number) != len(guess_number):
    print(f'Enter {len(original_number)} digit number:')
    continue

  if(len(guess_number) != len(set(guess_number))):
    print('Duplicate numbers are not allowed!')
    continue

  if int(original_number) - int(guess_number) == 0:
    print('Fermi ' * len(original_number))
    print('YOU WON!!!')
    break

  for i in range(len(original_number)):
    for j in range(len(guess_number)):
      if original_number[i] == guess_number[j]:
        if i == j:
          output.append('Fermi')
        else:
          output.append('Pico')

  output_string = ''
  for items in output:
    output_string = output_string + ' ' + items
  print(output_string)

  if len(output) == 0:
    print('Bagel')
  else:
    print(output_string)
```
